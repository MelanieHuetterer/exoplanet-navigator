```python
def calculate_limiting_distance(d, ps):
    return 15 * (d / 6) / ps

#print(calculate_limiting_distance(6, 1))
```

```python
diameter = 6
df['es_max'] = calculate_limiting_distance(diameter, df.pl_orbsmax)
df.head()
```

```python
rows = len(df[df.sy_dist <= df.es_max])
print("Rows count: ", rows)
```

```python
filt = (df.sy_dist <= df.es_max)
df2 = df.loc[filt]
df2.head(10)
```

```python
#SNR0 [(R* RP (D/6)) / ((ES/10) PS)]2
def calculate_snr(rs, rp, d, es, ps):
    return ((rs*rp*(d/6))/((es/10)*ps))*2
```

```python
diameter = 6
df['snr'] = calculate_snr(df.st_rad, df.pl_rade, diameter, df. sy_dist, df.pl_orbsmax)
df.head()
```

```python
rows2 = len(df[df.snr > 5])
print("Rows count: ", rows2)
```

```python
filt = (df.snr > 5)
df3 = df.loc[filt]
df3.head(10)
```
